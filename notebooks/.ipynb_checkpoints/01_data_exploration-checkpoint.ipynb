```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

sns.set_theme(style='whitegrid', font_scale=1.2)
%matplotlib inline
```

## 1. Load & basic stats

```python
df = pd.read_parquet(PARQUET)
print(f'Shape: {df.shape}')
print(f'Columns: {df.columns.tolist()}')
```

```python
print('--- pIC50 stats (all rows) ---')
print(df['pIC50'].describe())

print(f'\nUnique proteins: {df["protein"].nunique():,}')
print(f'Unique SMILES:   {df["SMILES"].nunique():,}')
print(f'Unique entry_ids:{df["entry_id"].nunique():,}')
```

## 2. Filtering impact

```python
print('Rows passing all PoseBusters:', df['all_passed'].sum(), 
      f'({100*df["all_passed"].mean():.1f}%)')
print()
print('Assay type breakdown:')
print(df['assay'].value_counts())
print()
print('Rows with biochemical assay + all_passed:')
mask = (df['all_passed'] == True) & (df['assay'] == 'biochem')
print(f'  {mask.sum():,} rows')
print(f'  {df[mask]["protein"].nunique():,} unique proteins')
print(f'  {df[mask]["SMILES"].nunique():,} unique SMILES')
```

## 3. pIC50 distribution by assay type

```python
fig, axes = plt.subplots(1, 2, figsize=(14, 5))

# All data
df['pIC50'].hist(bins=80, ax=axes[0], color='steelblue', edgecolor='none')
axes[0].set_title('pIC50 distribution (all rows)')
axes[0].set_xlabel('pIC50')
axes[0].set_ylabel('Count')

# By assay type
for assay, group in df.groupby('assay'):
    group['pIC50'].hist(bins=60, ax=axes[1], alpha=0.5, label=assay, edgecolor='none')
axes[1].set_title('pIC50 by assay type')
axes[1].set_xlabel('pIC50')
axes[1].legend()

plt.tight_layout()
plt.show()
```

## 4. Protein family breakdown

```python
family_counts = df.drop_duplicates('protein')['family'].value_counts()
family_counts.plot(kind='bar', figsize=(10, 4), color='steelblue')
plt.title('Unique proteins per family')
plt.xlabel('Family')
plt.ylabel('Count')
plt.xticks(rotation=30, ha='right')
plt.tight_layout()
plt.show()
```

## 5. Boltz confidence metrics vs pIC50

```python
conf_cols = ['iptm', 'ptm', 'complex_iplddt', 'interaction_ptm', 'confidence_score']
sample = df[df['all_passed'] == True].sample(min(50_000, len(df)), random_state=42)

fig, axes = plt.subplots(1, len(conf_cols), figsize=(18, 4))
for ax, col in zip(axes, conf_cols):
    ax.scatter(sample[col], sample['pIC50'], alpha=0.05, s=2, color='steelblue')
    corr = sample[[col, 'pIC50']].corr(method='spearman').iloc[0, 1]
    ax.set_title(f'{col}\nSpearman r={corr:.3f}')
    ax.set_xlabel(col)
    ax.set_ylabel('pIC50')

plt.suptitle('Boltz confidence metrics vs pIC50', y=1.02)
plt.tight_layout()
plt.show()
```

## 6. Quick SMILES validity check

```python
from rdkit import Chem

sample_smiles = df['SMILES'].dropna().sample(10_000, random_state=42)
invalid = sample_smiles.apply(lambda s: Chem.MolFromSmiles(s) is None).sum()
print(f'Invalid SMILES in 10k sample: {invalid} ({100*invalid/10000:.2f}%)')
```

## 7. Sequence length distribution

```python
unique_seqs = df.drop_duplicates('protein')[['protein', 'sequence']].copy()
unique_seqs['seq_len'] = unique_seqs['sequence'].str.len()

unique_seqs['seq_len'].hist(bins=60, color='steelblue', edgecolor='none', figsize=(10, 4))
plt.axvline(1022, color='red', linestyle='--', label='ESM2 max (1022)')
plt.title('Protein sequence length distribution')
plt.xlabel('Sequence length (aa)')
plt.ylabel('Count')
plt.legend()
plt.tight_layout()
plt.show()

over_limit = (unique_seqs['seq_len'] > 1022).sum()
print(f'Proteins longer than ESM2 limit (1022 aa): {over_limit} '
      f'({100*over_limit/len(unique_seqs):.1f}%) - will be truncated')
```
